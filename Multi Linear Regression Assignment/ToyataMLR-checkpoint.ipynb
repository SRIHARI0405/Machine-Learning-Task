```python
#Model Deletion Dignostics:
#Detecting influencers/outliers
#Using Cook's Distance
model_influencers = model.get_influence()
(a,_) = model_influencers.cooks_distance
```

```python
#Plot the influencers values using stem plot
plt.subplots(figsize = (16,4))
plt.stem(np.arange(len(df)),np.round(a,3))
plt.title('Influencers Plot')
plt.xlabel('Raw index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
#Index and values of influencer
(np.argmax(a), np.max(a))
```

```python
#multicolliniraity issue not exist
X= df2.iloc[:,1:8]
Y=df2["Price"]
```

```python
#Data Partition
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.5,random_state = 42)
```

```python
X_train.shape
X
```

```python
#model fitting
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(X_train,Y_train)
```

```python
Y_pred_train = LR.predict(X_train)
Y_pred_test = LR.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
print('Training Error',np.sqrt(mean_squared_error(Y_pred_train,Y_train)).round(2))
print('Testing Error',np.sqrt(mean_squared_error(Y_pred_test,Y_test)).round(2))
```
