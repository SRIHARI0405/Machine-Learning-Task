```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#load the dataset
df = pd.read_csv('wine.csv')
df
```

```python
#Data transformation
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
X = df.iloc[:,1:]
X_scale = SS.fit_transform(X)
```

```python
#PCA 
from sklearn.decomposition import PCA
pca = PCA(n_components=(3))
pc = pca.fit_transform(X_scale)
pc_value = pd.DataFrame(pc,columns=['PC1','PC2','PC3'])
```

```python
#hierarchial clustering
import matplotlib.pyplot as plt
plt.scatter(pc_value['PC1'],pc_value['PC2'])
```

```python
import scipy.cluster.hierarchy as shc
single_dend = shc.dendrogram(shc.linkage(pc_value,method='single'))
complete_dend = shc.dendrogram(shc.linkage(pc_value,method='complete'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='single')
cluster = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
Y = cluster.fit_predict(pc_value)
Y = pd.DataFrame(Y)
Y.value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(pc_value['PC1'],pc_value['PC2'],c=cluster.labels_, cmap='rainbow')
```

```python
#kmeans
from sklearn.cluster import KMeans
KM = KMeans(n_clusters=3)
Y=KM.fit_predict(pc_value)
Y = pd.DataFrame(Y)
Y.value_counts()
```

```python
k_value=[]
for i in range(2,10):
    KM = KMeans(n_clusters=i)
    Y=KM.fit_predict(pc_value)
    k_value.append(KM.inertia_)
print(k_value)
#elbow cure
import matplotlib.pyplot as plt
plt.scatter(range(2,10),k_value)
plt.plot(range(2,10),k_value)
```

```python
#kmeans
from sklearn.cluster import KMeans
KM = KMeans(n_clusters=4)
Y=KM.fit_predict(pc_value)
Y = pd.DataFrame(Y)
Y.value_counts()
```

```python
#DBSCAN
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=2 , min_samples=3) #eps = epsalon = radius
dbscan.fit(X_scale)
y=dbscan.labels_
Y=pd.DataFrame(y)
Y.value_counts()
```
