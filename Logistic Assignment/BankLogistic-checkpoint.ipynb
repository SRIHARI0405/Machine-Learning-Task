```python
#seperate the continus and categorical
df_object = df.select_dtypes('object')
df_numerical = df.select_dtypes('int64')
```

```python
#data transformation for categorical
from sklearn.preprocessing import LabelEncoder
for i in df_object.iloc[:,:-1]:
    LE = LabelEncoder()
    df_object[i] = LE.fit_transform(df_object[i])
```

```python
#combine the categorical and continuous datatype
df_new_value = pd.concat([df_numerical,df_object],axis = 1)
```

```python
#Data Transformation
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
df_new_value = SS.fit_transform(df_new_value.iloc[:,:-1])    
```

```python
df_new_value = pd.DataFrame(df_new_value)
df_new_value
```

```python
#Data Divison
X = df_new_value
Y = pd.DataFrame(df.iloc[:,-1])
```

```python
#Data Partition
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size =0.3,random_state = 42)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
#Data model
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train,Y_train)
LR.score(X_train,Y_train)
```

```python
Y_pred_train = LR.predict(X_train)
Y_pred_test = LR.predict(X_test)
```

```python
#Accuracy metrics
from sklearn.metrics import accuracy_score,classification_report
print('Training Accuracy',accuracy_score(Y_pred_train,Y_train).round(2))
print('Testing Accuracy',accuracy_score(Y_pred_test,Y_test).round(2))
```

```python
result = classification_report(Y_pred_train,Y_train)
print(result)
```

```python
#cross validation for logisitic regression
from sklearn.model_selection import KFold,cross_val_score
K = KFold(n_splits = 5)
result = cross_val_score(LR,X,Y,cv = 5)
print(result)
```

```python
print(np.mean(result).round(2))
```
