```python
#load the train dataset
df = pd.read_csv('SalaryData_Train.csv')
df
```

```python
#print frist five rows
df.head()
```

```python
#print the columns
df.columns
```

```python
#print the datatypes
df.dtypes
```

```python
#Data be divided into continous and categorical
df_object = df.select_dtypes('object')
df_numerical = df.select_dtypes('int64')
```

```python
#Data transformation for train data.
from sklearn.preprocessing import LabelEncoder
for i in df_object.iloc[:,:-1]:
    LE = LabelEncoder()
    df_object[i] = LE.fit_transform(df_object[i])    
```

```python
#combine the categorical and numerical datatypes
df_new_train = pd.concat([df_numerical,df_object],axis =1)
df_new_train
```

```python
#split your train data into x and y
X_train = df_new_train.iloc[:,:-1]
Y_train = df_new_train.iloc[:,-1]
```

```python
print(X_train.shape,Y_train.shape)
```

```python
#Load the test data
df_test = pd.read_csv('SalaryData_Test.csv')
df_test
```

```python
#print the frist five rows of test data
df_test.head()
```

```python
#print the columns
df_test.columns
```

```python
#print the datatypes
df_test.dtypes
```

```python
#sepearte the catergorical and continous data
df_test_object = df_test.select_dtypes('object')
df_test_numerical = df_test.select_dtypes('int64') 
```

```python
#Data Transformation for testing data 
from sklearn.preprocessing import LabelEncoder
for i in df_test_object.iloc[:,:-1]:
    LE = LabelEncoder()
    df_test_object[i] = LE.fit_transform(df_test_object[i])
```

```python
#combine catergorical and continous data
df_new_test = pd.concat([df_test_numerical,df_test_object],axis = 1)
df_new_test
```

```python
#Data division for test data
X_test = df_new_test.iloc[:,:-1]
Y_test = df_new_test.iloc[:,-1]
```

```python
print(X_test.shape,Y_test.shape)
```

```python
#Data model
from sklearn.naive_bayes import MultinomialNB
MNB = MultinomialNB()
MNB.fit(X_train,Y_train)
MNB.score(X_train,Y_train)
```

```python
Y_pred_train = MNB.predict(X_train)
Y_pred_test = MNB.predict(X_test)
```

```python
#Accurarcy metrics
from sklearn.metrics import accuracy_score
print('Training Accuracy',accuracy_score(Y_pred_train,Y_train).round(2))
print('Testing Accuracy',accuracy_score(Y_pred_test,Y_test).round(2))
```
