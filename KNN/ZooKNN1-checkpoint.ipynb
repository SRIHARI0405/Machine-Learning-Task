```python
#Data transformation for object
from sklearn.preprocessing import LabelEncoder
for i in df['animal name']:
    LE = LabelEncoder()
    df['animal name'] = LE.fit_transform(df['animal name'])
```

```python
#dependent and Independent Variable
X = df.iloc[:,:-1]
Y = pd.DataFrame(df['type'])
```

```python
#Data Transformation
from sklearn.preprocessing import StandardScaler
for i in X:
    SS = StandardScaler()
    X = SS.fit_transform(X)
```

```python
X = pd.DataFrame(X,columns = ['animal name', 'hair', 'feathers', 'eggs', 'milk', 'airborne',
       'aquatic', 'predator', 'toothed', 'backbone', 'breathes', 'venomous',
       'fins', 'legs', 'tail', 'domestic', 'catsize'])
```

```python
X
```

```python
#Data Partition
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3,random_state = 42)
```

```python
#Data modeling
from sklearn.neighbors import KNeighborsClassifier
KNN  = KNeighborsClassifier(n_neighbors = 6)
KNN.fit(X_train,Y_train)
KNN.score(X_train,Y_train).round(2)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
Y_pred_train = KNN.predict(X_train)
Y_pred_test = KNN.predict(X_test)
```

```python
#metrics
from sklearn.metrics import accuracy_score
print('Training Accuracy',accuracy_score(Y_pred_train,Y_train).round(2))
print('Testing Accuracy',accuracy_score(Y_pred_test,Y_test).round(2))
```

```python
#cross validation
from sklearn.model_selection import KFold,cross_val_score
K = KFold(n_splits = 5)
result = cross_val_score(KNN,X,Y,cv =6)
print(result)
```

```python
#mean of the cross validation
print(np.mean(result).round(2))
```
