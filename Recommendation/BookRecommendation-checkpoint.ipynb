```python
df_book = book.pivot_table(index="User.ID",columns="Book.Title",values="Book.Rating")
```

```python
df_book.fillna(0,inplace=True)
df_book
```

```python
from scipy.spatial.distance import cosine,correlation
```

```python
from sklearn.metrics import pairwise_distances
```

```python
df_books=1-pairwise_distances(df_book.values,metric='cosine')
```

```python
df_new_books =pd.DataFrame(df_books)
```

```python
def give_recommendation(user_id):
    tem = list(df_new_books.sort_values([user_id], ascending= False).head().index)
    books_list = []
    
    for i in tem:
        books_list = books_list + list(book[book['User.ID']==i]['Book.Title'])    
    return set(books_list)-set(book[book['User.ID']==user_id]['Book.Title'])
```

```python
give_recommendation(12)
```
