```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#load the tarin dataset
df_train = pd.read_csv('SalaryData_Train(1).csv')
df_train
```

```python
#check the shape of train data
df_train.shape
```

```python
#check the columns of train data
df_train.columns
```

```python
#check the datatype of train data
df_train.dtypes
```

```python
#check the null values in train data
df_train.isnull().sum()
```

```python
#seperate the data type 
df_train_object = df_train.select_dtypes('object')
df_train_numerical = df_train.select_dtypes('int64')
```

```python
#Data Transformation for object
from sklearn.preprocessing import LabelEncoder
for i in df_train_object:
    LE = LabelEncoder()
    df_train_object[i] = LE.fit_transform(df_train_object[i])
```

```python
#combine the object and numerical datatype
df_train_new = pd.concat([df_train_numerical,df_train_object],axis = 1,ignore_index = True)
```

```python
#Data Transformation for train data
from sklearn.preprocessing import StandardScaler
for i in df_train_new:
    SS = StandardScaler()
    df_train_new = SS.fit_transform(df_train_new)
```

```python
df_train_new = pd.DataFrame(df_train_new,columns =['age', 'workclass', 'education', 'educationno', 'maritalstatus',
       'occupation', 'relationship', 'race', 'sex', 'capitalgain',
       'capitalloss', 'hoursperweek', 'native', 'Salary'])
```

```python
df_train_new
```

```python
#Here are the training data
X_train = df_train_new.iloc[:,0:13][0:200]
Y_train = pd.DataFrame(df_train.iloc[:,-1])[0:200]
```

```python
print(X_train,Y_train)
```

```python
#load the testdata set
df_test = pd.read_csv('SalaryData_Test(1).csv')
df_test
```

```python
#check the shape of train data
df_test.shape
```

```python
#check the columns of train data
df_test.columns
```

```python
#check the datatype of train data
df_test.dtypes
```

```python
#check the null values in train data
df_test.isnull().sum()
```

```python
#seperate the data type 
df_test_object = df_test.select_dtypes('object')
df_test_numerical = df_test.select_dtypes('int64')
```

```python
#Data Transformation for object
from sklearn.preprocessing import LabelEncoder
for i in df_test_object:
    LE = LabelEncoder()
    df_test_object[i] = LE.fit_transform(df_test_object[i])
```

```python
#combine the object and numerical datatype
df_test_new2 = pd.concat([df_test_numerical,df_test_object],axis = 1,ignore_index = True)
```

```python
#Data Transformation
from sklearn.preprocessing import StandardScaler
for i in df_test_new2:
    SS = StandardScaler()
    df_test_new2 = SS.fit_transform(df_test_new2)
```

```python
df_test_new3 = pd.DataFrame(df_test_new2,columns = ['age', 'workclass', 'education', 'educationno', 'maritalstatus',
       'occupation', 'relationship', 'race', 'sex', 'capitalgain',
       'capitalloss', 'hoursperweek', 'native', 'Salary'])
df_test_new3
```

```python
X_test = df_test_new3.iloc[:,0:13][0:200]
Y_test = pd.DataFrame(df_test.iloc[:,13])[0:200]
```

```python
print(X_train.shape,Y_train.shape,X_test.shape,Y_test.shape)
```

```python
#Data Modeling
from sklearn.svm import SVC
S = SVC()
S.fit(X_train,Y_train)
S.score(X_train,Y_train)
```

```python
Y_pred_train = S.predict(X_train)
Y_pred_test = S.predict(X_test)
```

```python
#metrics
from sklearn.metrics import accuracy_score
print('Training Accuracy',accuracy_score(Y_pred_train,Y_train).round(2))
print('Testing Accuracy',accuracy_score(Y_pred_test,Y_test).round(2))
```

```python
#Doing the hyper parameter tuning for getting better Accuracy
from sklearn.model_selection import GridSearchCV
kernel = ['linear','ploy','rbf','sigmoid']
C = [1,3,5,7,9]
gamma = [0.01,0.1,0.2,0.3,0.5]
param = {
    'kernel':kernel,
    'C':C,
    'gamma':gamma,
}
result = GridSearchCV(estimator = S,param_grid = param,n_jobs =-1,cv =5)
result.fit(X_train,Y_train)
print(result)
```

```python
#checking the best parameter
result.best_params_
```

```python
#Data Modeling
from sklearn.svm import SVC
S = SVC(kernel ='linear',C = 5, gamma = 0.01)
S.fit(X_train,Y_train)
S.score(X_train,Y_train)
```

```python
Y_pred_train = S.predict(X_train)
Y_pred_test = S.predict(X_test)
```

```python
#metrics
from sklearn.metrics import accuracy_score
print('Training Accuracy',accuracy_score(Y_pred_train,Y_train).round(2))
print('Testing Accuracy',accuracy_score(Y_pred_test,Y_test).round(2))
```

```python
#with the help of optimazation technique  like gridsearch cv i got kernal as linear C as 5 and gamma as 0.01 and getting better testing accuracy compares to the previous one.
```

```python
#cross validation
from sklearn.model_selection import KFold,cross_val_score
X = pd.concat([X_train,X_test],axis = 0)
Y = pd.concat([Y_train,Y_test],axis = 0)
K = KFold(n_splits = 5)
result = cross_val_score(S,X,Y,cv = K)
print(result)
```
